```julia
# filename: test_script.jl
using Printf
using DelimitedFiles

# readlines
x = Float64[]
y = Float64[]
```

```julia
open("decibel.dat", "r") do f
  for (i,line) in enumerate(eachline(f))
    println(line)
    a , b = split(line, ",")
    println(a, "\t", b)
    push!(x, parse(Float64, a))
    push!(y, parse(Float64, b))
  end
end
```

```julia
# read columns to arrays
arr = readdlm("decibel.dat", ',', header=false, skipstart=0, skipblanks=true)
x = arr[:,1]
y = arr[:,2]
```

```julia
for i in (1:length(x))
  @printf("i= %d, X= %f, Y= %f \n", i, x[i], y[i])
end
```
