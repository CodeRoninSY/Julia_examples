```julia
using Plots
```

```julia
x = -3:0.1:3
f(x) = x^2
y = f.(x)
```

```julia
# load GR backend
gr() 
```

```julia
plot(x, y, label="inline")
# "scatter!" mutates the plot, plotting regenerates after plot and scatter data is added
scatter!(x, y, label="points")
```
