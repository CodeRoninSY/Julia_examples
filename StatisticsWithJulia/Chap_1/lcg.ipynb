```julia
using Plots; pyplot()
```

```julia
a, c, m = 69069, 1, 2^32
```

```julia
next(z) = (a*z + c) % m
```

```julia
N = 10^6
```

```julia
data = Array{Float64,1}(undef, N)
```

```julia
x = 808
```

```julia
for i in 1:N
    data[i] = x/m
    global x = next(x)
end
```

```julia
p1 = scatter(1:1000, data[1:1000], c=:blue, m=4, msw=0)
p2 = histogram(data, bins=50, normed=:true, ylims=(0,1.1))
plot(p1, p2, size=(800, 400), legend=:none)
```
